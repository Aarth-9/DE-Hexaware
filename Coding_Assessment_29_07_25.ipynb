```python
df1 = pd.read_csv('/content/drive/My Drive/datasets/product_prices.csv')
```

```python
df1.head()
```

```python
print("Sales Data Columns:", df.columns.tolist())
print("\nProduct Prices Columns:", df1.columns.tolist())
```

```python
inner_join = pd.merge(df, df1, left_on='prod_id', right_on='product_id', how='inner')
print("Length after Inner join: ", len(inner_join))
print("\nInner Join:\n", inner_join.head())
```

```python
left_join = pd.merge(df, df1, left_on='prod_id', right_on='product_id', how='left')
print("Length after Left join: ", len(left_join))
print("\nLeft Join:\n", left_join.head())
```

```python
right_join = pd.merge(df, df1, left_on='prod_id', right_on='product_id', how='right')
print("Length after Right join: ", len(right_join))
print("\nRight Join:\n", right_join.head())
```

```python
outer_join = pd.merge(df, df1, left_on='prod_id', right_on='product_id', how='outer')
print("Length after outer join: ", len(outer_join))
print("\nOuter Join:\n", outer_join.head())
```
